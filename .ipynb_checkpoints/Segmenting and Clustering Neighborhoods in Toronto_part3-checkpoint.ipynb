# Segmenting and Clustering Neighborhoods in Toronto

## Question 3

```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium
```

### Importing the dataset from part 2

```python
df = pd.read_csv('Part_2.csv')
df.head()
```

### Exploring the dataset

#### Use geopy library to get the latitude and longitude values of Toronto, Ontario Canada.

```python
address = 'Toronto, Ontario Canada'

geolocator = Nominatim(user_agent="Toronto_Neigh")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

#### Create a map of Toronto with neighborhoods superimposed on top.

Creating a map of Toronto for all Neighborhoods.

```python
# create map of Toronto including all Neighborhoods in the datset using latitude and longitude values
map = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)  
    
map
```

Keeping only those rows in dataset which contain 'Toronto' in the Borough column.

```python
df_toronto = df[df['Borough'].str.contains('Toronto')].reset_index(drop=True)
df_toronto.head()
```

```python
# create map of Toronto including only Neighborhoods in the datset where Borough column has 'Toronto' in it 
#using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
